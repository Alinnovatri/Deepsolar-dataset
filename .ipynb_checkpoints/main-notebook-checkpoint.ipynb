En este histograma se muestra la cantidad de paneles solares. La mayoría de ellos son inferiores a 50 , esto se debe a que gran parte de ellos son para uso residencial por lo que no se requiere de una gran cantidad, se ve que coforme aumenta la cantidad de paneles, disminuye la cantidad de sitios que lo ocupan, esto de debe a un uso que ya no corresponde al residencial, sino a sitios que se encarga para la generación un para un uso comercial.

### Representación por sector

```python
datadeepsolar.showBarGraphSector()
```

En esta gráfica de barras se comparar la cantidad de paneles solares utilizados por el sector residencia y no residencialm. En el primer conjunto es representado por la cantidad de paneles solares y el segundo de la cantidad de sistemas solares. Al sacar un promedio de paneles, se encuentra en aproximadamente en una zona de los Estados Unidos(referente a una localidad) hay aproximadamente entre 25 y 35 paneles solares y por sistemas solares 1.5. Al representarlo por sector se muestra claramemte que el sector residencial es con el que cuenta con mayor cantidad de paneles y sistemas fotovoltaicos cabe recalcar que el uso de paneles solares para fines no residenciales incluye usos comerciales, industriales y de transporte de energía solar.

Existe una gran cantidad de paneles fotovoltaicos en zonas residenciales que los que son utilizados por el sector comercial, industrial y de transporte, Aunque existen pocos sistemas fotovoltaicos con fines no residenciales que residenciales, esto implica que son más hogares en los Estados Unidos que tiene paneles fotovoltaicos que empresas, industrias y corporaciones por lo que debera incentivar a este sector.
